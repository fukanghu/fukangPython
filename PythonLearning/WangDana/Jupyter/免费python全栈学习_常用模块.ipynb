```python
# datetime.timedelta
# è¡¨ç¤ºä¸€ä¸ªæ—¶é—´é—´éš”

from datetime import datetime, timedelta

t1 = datetime.now()

print(t1.strftime("%Y-%m-%d %H:%M:%S"))
# tdè¡¨ç¤ºä»¥å°æ—¶çš„æ—¶é—´é•¿åº¦
td = timedelta(hours=24)
# å½“å‰æ—¶é—´åŠ ä¸Šæ—¶é—´é—´éš”åï¼ŒæŠŠå¾—åˆ°çš„çš„æ˜¯æ—¶é—´æ ¼å¼åŒ–è¾“å‡º
print((t1+td).strftime("%Y-%m-%d %H:%M:%S"))
```

```python
# timeitï¼šæ—¶é—´æµ‹é‡å·¥å…·

# æµ‹é‡ç¨‹åºè¿è¡Œæ—¶é—´é—´éš”å®éªŒ
def p():
    time.sleep(3.6)
    
t1 = time.time()
p()
print(time.time() - t1)
```

```python
import timeit
# ç”Ÿæˆåˆ—è¡¨ä¸¤ç§æ–¹æ³•çš„æ¯”è¾ƒ
# å¦‚æœå•çº¯æ¯”è¾ƒç”Ÿæˆä¸€ä¸ªåˆ—è¡¨çš„æ—¶é—´ï¼Œå¯èƒ½å¾ˆéš¾å®ç°
c = '''
sum = []
for i in range(1000):
    sum.append(i)
'''
# åˆ©ç”¨timeitè°ƒç”¨ä»£ç ï¼Œæ‰§è¡Œ100000æ¬¡ï¼ŒæŸ¥çœ‹è¿è¡Œæ—¶é—´
t1 = timeit.timeit(stmt="[i for i in range(1000)]", number=100000) 
# æµ‹é‡ä»£ç cæ‰§è¡Œ100000æ¬¡è¿è¡Œç»“æœ
t2 = timeit.timeit(stmt=c, number=100000)
print(t1)
print(t2)
```

```python
# timeitï¼šå¯ä»¥æ‰§è¡Œä¸€ä¸ªå‡½æ•°ï¼Œæ¥æµ‹é‡ä¸€ä¸ªå‡½æ•°çš„æ‰§è¡Œæ—¶é—´
def doIt():
    num = 3
    for i in range(num):
        print("Rpeat for {0}".format(i))
# æ‰§è¡Œå‡½æ•°ï¼Œé‡å¤åæ¬¡
t = timeit.timeit(stmt=doIt, number=10)
print(t)
```

```python
s = '''
def doIt(sum):
    for i in range(num):
        print("Rpeat for {0}".format(i))
'''
# æ‰§è¡ŒdoIt(num)
# setupè´Ÿè´£æŠŠç¯å¢ƒå˜é‡å‡†å¤‡å¥½
# å®é™…ä¸Šç›¸å½“äºç»™timeitåˆ›é€ äº†ä¸€ä¸ªå°ç¯å¢ƒ
# åœ¨åˆ›å»ºçš„å°ç¯å¢ƒä¸­ï¼Œä»£ç æ‰§è¡Œçš„é¡ºåºå¤§è‡´æ˜¯
# def doIt(num):
# ......
# num = 3
# doIt(sum)
t = timeit.timeit("doIt(num)", setup=s+"num=3", number=10)
print(t)
```

```python

# choice() éšæœºè¿”å›åºåˆ—ä¸­çš„æŸä¸ªå€¼
#  æ ¼å¼ï¼šrandom.choice(åºåˆ—)
#  è¿”å›å€¼ï¼šåºåˆ—ä¸­çš„æŸä¸ªå€¼

l = [str(i)+"haha" for i in range(10)]
print(l)
rst = random.choice(l)
print(rst)
```
