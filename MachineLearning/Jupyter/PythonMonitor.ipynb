# 数据获取
## urllib的基本使用


```python
from urllib import request

url = "http://www.baidu.com"
res = request.urlopen(url) # 获取响应

print(res.info()) # 响应头
print(res.getcode()) # 状态码, 2xx, 3xx, 4xx, 5xx
print(res.geturl()) # 返回响应地址
```

```python
html = res.read()
# print(html)
html = html.decode("utf-8")
print(html)
```

```python
# 添加header信息，最基本反扒措施
url = "http://www.dianping.com"
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36"
}
req = request.Request(url, headers=header)
res = request.urlopen(req) # 获取响应

print(res.info()) # 响应头
print(res.getcode()) # 状态码, 2xx, 3xx, 4xx, 5xx
print(res.geturl()) # 返回响应地址
```

# beautifulsoup 的简单应用

```python
from bs4 import BeautifulSoup
import requests

url = "http://wsjkw.sc.gov.cn/scwsjkw/gzbd/fyzt.shtml"
res = requests.get(url)
res.encoding = "utf-8"
html = res.text
soup = BeautifulSoup(html)
soup.find("h2").text

a = soup.find("a")
print(a)
print(a.attrs)
print(a.attrs["href"])
```

# 使用re解析

```python
import re
text = context.text
# print(text)
patern = "新增确诊病例(\d+)例"
res = re.search(patern, text)
print(res)
```

```python
print(res.groups())
print(res.group(0))
print(res.group(1))
```

```python
patern = "新增确诊病例(\d+).*?出院病例(\d+).*?疑似病例(\d+).*?死亡病例(\d+)" # .* 后面叫做非贪婪匹配
res = re.search(patern, text)
print(res.groups())
print(res.group(0))
print(res.group(1), res.group(2), res.group(3), res.group(4))
```
